```python
sentence ="""The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France. It is named after the engineer Gustave Eiffel, whose company designed and built the tower. Constructed from 1887 to 1889 as the entrance to the 1889 World's Fair, it was initially criticized by some of France's leading artists and intellectuals for its design, but it has become a global cultural icon of France and one of the most recognizable structures in the world. The Eiffel Tower is the most-visited paid monument in the world; 6.91 million people ascended it in 2015.
"""

# Person Eg: Aditya Raj 
# Place or LOcation Eg: India
# Date Eg: September, 24-09-1989
# TIme Eg: 4:30 pm
# Money Eg: 1 million dollar
# Organization Eg: iNeuron Private Limited
# Percent Eg: 20%, Twenty Percent

```

```python
import nltk
words = nltk.word_tokenize(sentence) # We have divided the sentences into words and saved their list in words.
print(words) 
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
tag_elements = nltk.pos_tag(words) # POS tagging identifies which words are nouns/proper nouns, helping NER accurately recognize entities like people, places, and organizations.
print(tag_elements)
```
