Wordnet Lemmatizer
Lemmatization => this technique is also like stemming. The output we will get after lemmatization is called 'lemma', which is a root word rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word taht means the same thing. 

NLTK proved WordNetLemmatizer class which is a thin wrapper around wordnet corpus. this class uses morphy() function to the WordNet CorpusReader class to a find a lemma. let us understand it with an example-

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
import nltk

nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
"""
POS-Noun-n,
verb-v,
adjective-a,
adverb-r
"""
print(lemmatizer.lemmatize('going')) # here pos is n, going gets treeated like a noun.
print(lemmatizer.lemmatize('going', pos='v')) # here pos is v, and going is treated like a verb.
```

```python
words = ["writing", "eating", "eats", "eaten", "writes", "programming", "programs", "history", "finally", "finalize"]
```

```python
for word in words:
    print(word+ "--->" +lemmatizer.lemmatize(word, pos='v'))
```
