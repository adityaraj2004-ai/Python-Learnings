Stemming => It is the process of reducing a word to its word stem affixing any of the prefixes or suffixes or to the roots of the word known as a lemma. It is important in NLU and NLP.

Classification problem
We have comments of a product and we have to classify them as -> Positive Review and Negative Review

Review consists of words like -> [eating, eaten, eat] => eat(Root Word), [going gone goes] => go (Root Word)


```python
words = ["writing", "eating", "eats", "eaten", "writes", "programming", "programs", "history", "finally", "finalize"]
```

```python
# 1) Stemming technique => PorterStemmer

from nltk.stem import PorterStemmer
stemming = PorterStemmer()
```

```python
for word in words:
    print(word + "----->" +stemming.stem(word))
```

```python
stemming.stem("congratulation")
```

```python
stemming.stem("sitting")
```

```python
# RegexpStemmer
from nltk.stem import RegexpStemmer

reg_stemmer = RegexpStemmer('ing|s$|e$|able$', min=4) # here dollar after the words means last bit, if we remove dollar then it will remove the given words as prefixes also
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

```python
# Snowball stemmer => more accurate then the above two.

from nltk.stem import SnowballStemmer
snowballstemmer = SnowballStemmer('english')
```

```python
for word in words:
    print(word + "--->" +snowballstemmer.stem(word))
```

```python
#comparison between Porter Stemming and SnowBall Stemming
# Even if snowball stemming performs excetionally well, yet it does not perform on some words like hsitory
print(stemming.stem('sportingly'),stemming.stem('fairly'))
print(snowballstemmer.stem('sportingly'),snowballstemmer.stem('fairly'))
```

```python
# one word where both Porter stemming and snowballStemming fails
print(stemming.stem('goes'))
print(snowballstemmer.stem('goes'))
```

All the above issues we have seen where our stemming techniques fails (that is why we cannot use stemming in chatbots) => are treated by lemmatization --> lemmatization has dictionary of all the root words, thus it performs better.
