```python
# Strings

str = "ADITYA RAJ" #  In Python we can traverse each char of the String same as in arrays

for i in str:
    print(i)
```

```python
# While Loop: Continues to execute as long as the condition is true.

count = 5
while count < 10:
    print(count)
    count = count +1
```

```python
#Loop Control Statements 
# 1) break: The break statements exits the loop prematurely 

for i in range(5):
    if i == 5: # As soon as i becomes 5 the loop is exited
        break
    print(i)
```

```python
# 2) continue: The continue statement skips the current iteration and continues with next 

for i in range(10):
    if i % 2 == 0: # skips the iteration where i is divisible by 2.
        continue
    print(i)
```

```python
# 3) pass: The pass statement is a null operation, it does nothing, at that iteration but it goes through that iteration.It does not skip it.

for i in range(10):
    if i%2 == 0:
        pass # when we have to do nothing inside a statement block
    print(i)
  
        
     
    
```

```python
# Nested Loops
# a loop inside a loop

for i in range(5):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```
