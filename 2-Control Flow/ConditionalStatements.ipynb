```python
# else
# The else statement executes a block of code if the condition in the if statement is False.

age = 16

if age>=18:
    print("You are eligible for voting")
else:
    print("You are a minor")    
```

```python
# elif 
# allows you to check for multiple statements it stands for "else if"

age = 17

if age < 13:
    print("You are a child")
elif age < 18:
    print("You are a teenager")
else:
    print("Hey! You are an adult")        
```

```python
# Nested Conditionals
# one or more if elif conditionals inside each other 
# given number tell even, odd, negative 

num = int(input("Enter the number: "))
if num >=0:
    print("The number is Positive ") 
    if num % 2 == 0:
        print("THe number is also Even ")
    else:
        print("The number is also Odd")        
else:
    print("The number is negative")        
```

```python
# Practical Example
# Determine whether a year is leap year.

year = int(input("Enter the Year"))

if year < 0:
    print("year cannot be negative")
elif year>=0 & year %4 ==0:
    print("It is a Leap Year")
else:
    print("It is not a Leap Year")        
    
```
