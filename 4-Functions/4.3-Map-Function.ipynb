THE map()funtion in python 
The Map function applies a given function to all items in an input list(or any other iterative) and returns a map object(an iterator). This is particularyly useful for transforming data in a list comprehensively.


```python
# Defining a simple square function
def sq(x):
    return x**2

sq(10)
```

```python

numbers= [1, 2, 3, 4, 5]

# map takes, two parameter one is the function and other is thr iterable
result1=list(map(sq,numbers))
result2=list(map(lambda x:x**2, numbers))
print(result1)
print(result2)
```

```python
# Map multiple iterables
numbers1=[1,2,3]
numbers2=[4,5,6]

added_numbers=list(map(lambda x,y:x+y,numbers1,numbers2))
print(added_numbers)
```

```python
# map() to convert a list of strings into integers

str_list = ["1", "2", "3"]
# to_int = list(map(lambda x:int(x), str_list))

# easier way
to_int = list(map(int,str_list))
type = list(map(type,to_int))
print(type)
print(to_int)
```

```python
words = ['apple', 'banana', 'grapes']
upper_words=list(map(str.upper,words))

print(upper_words)
```

```python
# using map function witrh dictionary

def get_name(person):
    return person['name']

people=[
    {'name':'Aditya', 'age':30},
    {'name':'Raghav', 'age':30},    
]

list(map(get_name,people))
```
