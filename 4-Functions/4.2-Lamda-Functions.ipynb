Lamda Functions are anonymous Functions which are defined using lamda keyword. They can have any number of arguments but but only one expression. They are used for short operation or as arguments to higher order functions.

```python
#Syntax
lambda arguments: expression
```

```python
# Normal function
def addition(a,b):
    return a+b


result = addition(5,6)
print(result)


# Lamda Function

result2= lambda a,b:a+b
print(type(result2))
print(result2(5,6))

```

```python
even1 = lambda num:num%2==0

print(even1(12))
```

```python
def addition(x,y,z):
    return x+y+z

addition(2,3,4)
```

```python
addition1 = lambda x,y,z:x+y+z
addition1(2,3,4)
```

```python
# map()-> A short overview
numbers=[1,2,3,4,5,5,6,7,7,8]

list(map(lambda x: x**2,numbers))
```
