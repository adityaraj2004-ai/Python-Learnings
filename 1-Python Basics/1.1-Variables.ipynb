```python
# Dynamic typing- Python allows the type of variable to change as the program executes

var = 10
print(var, type(var)) #10 <class 'int'>

var = "Hello"
print(var, type(var)) #Hello <class 'str'>

#again dynamically changing the type at runtime
var = 6.11
print(var, type(var))


```

```python
# Creating a simple calculator for two numbers;

num1 = float(input("Enter the num1: "))
num2 = float(input("Enter the num2: "))

sum = num1+num2
difference = num1-num2
product = num1*num2
quotient = num1/num2

print("Sum: ", sum)
print("Difference: ", difference)
print("Product: ", product)
print("Quotient: ", quotient)
```
